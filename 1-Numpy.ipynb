```python
# dot

''' This function returns the product of two arrays. For 2-D vectors, it is the equivalent to matrix multiplication.
For 1-D arrays, it is the inner product of the vectors. For N- dimensional array, it is a sum product over the last axis of a and the second axis of b'''
```

```python
a = np.array([[1,2],[3,4]])
b = np.array([[11,12],[13,14]])

print("a :", a)
print()
print("b :", b)
print()
print("a x b =", np.dot(a,b))
```
